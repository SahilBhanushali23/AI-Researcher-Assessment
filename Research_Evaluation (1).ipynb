```python
#multi-turn conversations
from  google import  genai

client = genai.Client()

chat = client.chats.create(model="gemini-2.5-flash")

response = chat.send_message("A shipment carrying pharmaceutical products from Mumbai to Pune has been delayed by 3 hours due to highway congestion,analyze the situation and recommend the best immediate action.")
print(response.text)

print("-"*50)

response=chat.send_message("The customer has now informed us that the medicines are temperature-sensitive and must arrive within 8 hours,update your recommendation.")
print(response.text)
```

```python
from groq import Groq

client = Groq()

chat_completion = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    response_format={"type": "json_object"},
    messages=[
        {
            "role": "system",
            "content": "You are an AI logistics assistant."
        },
        {
            "role": "user",
            "content": """
A pharmaceutical shipment has been delayed due to heavy traffic.

Shipment Details:
- Source: Mumbai Distribution Center
- Destination: Pune Medical Warehouse
- Current Delay: 3 hours
- Cargo: Temperature-sensitive vaccines
- Maximum Allowed Delivery Time: 8 hours

Recommend the best alternative delivery route while minimizing delivery time and transportation cost.

Return the response in JSON format with:
- route
- eta
- cost
- reasoning
"""
        }
    ],
    temperature=0.2,
)

print(chat_completion.choices[0].message.content)
```

```python
from groq import Groq

client = Groq()

chat_completion = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {
            "role": "system",
            "content": "You are an AI logistics assistant specializing in route optimization."
        },
        {
            "role": "user",
            "content": """
A logistics company has two delivery routes for transporting pharmaceutical products.

Route A:
- Distance: 180 km
- Fuel Cost: ₹4,500
- Traffic: Moderate
- Delivery Time: 3.5 hours

Route B:
- Distance: 210 km
- Fuel Cost: ₹5,800
- Traffic: Heavy
- Delivery Time: 5 hours

The customer is a high-priority hospital that requires medicines as quickly as possible.

Explain why Route A is better than Route B considering fuel cost, traffic, and customer priority.
"""
        },
        {
            "role": "assistant",
            "content": """
Route A is the better choice because it has a lower fuel cost (₹4,500 vs ₹5,800), experiences only moderate traffic, and offers a significantly shorter delivery time of 3.5 hours compared to 5 hours. Since the customer is a high-priority hospital requiring urgent medicine delivery, Route A minimizes delays while also reducing transportation costs.
"""
        },
        {
            "role": "user",
            "content": """
Now summarize your recommendation into three concise bullet points.
"""
        }
    ],
    temperature=0.2,
)

print(chat_completion.choices[0].message.content)
```
